## Proje 2

**[16,21,11,8,12,22]**

* Yukarıdaki dizinin merge sort göre aşamalarını yazınız.
* Big-O gösterimini yazınız.

Merge sort, diziyi ikiye bölerek her bir yarıyı sıralar ve ardından bu yarıları birleştirir.

1. Diziyi ikiye böl:

* Sol yarı: [16, 21, 11]
* Sağ yarı: [8, 12, 22]

2. Sol yarıyı ikiye böl:

* Sol yarı: [16]
* Sağ yarı: [21, 11]

3. [21, 11] dizisini ikiye böl:

* Sol yarı: [21]
* Sağ yarı: [11]

4. [21] ve [11] dizilerini birleştir (sırala):

* [11, 21]

5. [16] ve [11, 21] dizilerini birleştir (sırala):

* [11, 16, 21]

6. Sağ yarıyı ikiye böl:

* Sol yarı: [8]
* Sağ yarı: [12, 22]

7. [12, 22] dizisini ikiye böl:

* Sol yarı: [12]
* Sağ yarı: [22]

8. [12] ve [22] dizilerini birleştir (sırala):

* [12, 22]

9. [8] ve [12, 22] dizilerini birleştir (sırala):

* [8, 12, 22]

10. Son olarak [11, 16, 21] ve [8, 12, 22] dizilerini birleştir (sırala):

* [8, 11, 12, 16, 21, 22]

### Merge Sort Big-O Gösterimi

* Best case (en iyi durum) O(n log n)

* Average case (ortalama durum) O(n log n)

* Worst case (en kötü durum) O(n log n)

* Merge sort, diziyi sürekli ikiye bölerek O(log n) derinlikte bölme işlemi yapar ve her birleştirme adımı O(n) zaman alır. Bu yüzden zaman karmaşıklığı O(n log n) olur.
